# Plottings based Anxiety Severity 

```python
sns.boxplot(x=data1.GAD_T, y=data1.Work, data=data1).set(title='Work Vs Anxiety Severity',xlabel= 'Anxiety Severity')
```

## Scoring GAD-7 Anxiety Severity

#### 0–4: minimal anxiety
#### 5–9: mild anxiety
#### 10–14: moderate anxiety
#### 15–21: severe anxiety

#### from this priliminary plot created form this dataset we can see that Unemployed gamers have a higher mean Anxiety compared to other categories also employed gamers have the least average.

```python
sns.boxplot(x=data1.SWL_T, y=data1.Work, data=data1).set(title=' Work Vs Satisfaction With Life',xlabel= 'Anxiety Severity')
```

## SWL - SATISFACTION WITH LIFE SCALE

#### 31 - 35 Extremely satisfied
#### 26 - 30 Satisfied
#### 21 - 25 Slightly satisfied
#### 20 Neutral
#### 15 - 19 Slightly dissatisfied
#### 10 - 14 Dissatisfied
#### 5 - 9 Extremely dissatisfied

#### Here again we can see that the "Satisfaction in life" is highest for Employed and lowest for unemployed Gamers.

```python
sns.boxplot(x=data1.SPIN_T, y=data1.Work, data=data1).set(title='Work Vs Social Phobia',xlabel= 'Anxiety Severity')
```

## Social Phobia Severity

#### 0 - 20 : VERY MILD
#### 21 - 30 : MILD
#### 31 - 40 : MODERATE
#### 41 - 50 : SEVERE
#### 50 or more : VERY SEVERE

#### Similar to the previous two plots we can again see that, unemployed gamers seem to have the highest average "Social Phobia", whereas employed gamers have the least.
